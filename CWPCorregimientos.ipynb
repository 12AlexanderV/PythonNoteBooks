```python
gpd.io.file.fiona.drvsupport.supported_drivers['KML'] = 'rw'
gdf = gpd.read_file('hhh.kml', driver='KML')

kml_file = path.join ('hhh.kml')

with open(kml_file, encoding = 'utf-8') as f:
    
  folder = parser.parse(f).getroot().Document.Folder
  

places = []
gf=  gdf.centroid
#gf = gpd.geodataframe(centroid)
x = gf.centroid.map(lambda p: p.x)
y = gf.centroid.map(lambda p: p.y)
#gd = gpd.GeoDataFrame(centroid)
#gdf["long"] = gdf.Center_point.map(lambda p: p.x)
#gdf["lat"] = gdf.Center_point.map(lambda p: p.y)
coord = pd.DataFrame(
    {
     'Longitude': x,
     'Latitude': y
     })



for place in folder.Placemark:
  data = {item.get("name"): item.text for item in
           place.ExtendedData.SchemaData.SimpleData}
  places.append({"PROVINCIA": data.get('PROVINCIA'),"DISTRITO": data.get('DISTRITO'),"CORREGIMIENTO": data.get('CORREGIMIENTO') })
                 #,"CENTROID":centroid})
                   #,
                  #"Province": data.get('Province'),
                  #"Comments": data.get('Comments'),
                  #"Coordinates": coords})


geo_desc = pd.DataFrame(places)

final_file = pd.concat([geo_desc, coord], axis = 1)

final_file.to_csv('2022-07-27_m2.csv', encoding = 'utf-8')

#print(pd)
#df = gpd.geodataframe(places)

#print(df)

#df = pd.DataFrame(gdf)
#print(df)
#df = df.set_index("DISTRITO")

#print(df)
#type(df)

#gdf['centroid'] = gdf.centroid

#gdf['centroid']




##df.bounds
##df.total_bounds

# bounds for individual geometries
#poly_geom = df.bounds
#b = poly_geom.apply(lambda row: box(row.minx, row.miny, row.maxx, row.maxy), axis=1)
#boxes = gpd.GeoDataFrame(poly_geom, geometry=b)
#print(boxes)

# visualize
#ax = df_centroids.plot()
#boxes.boundary.plot(color='b', ax=ax)
#plt.show()*/
```
